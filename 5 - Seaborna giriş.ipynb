```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings 
warnings.filterwarnings('ignore')
%matplotlib inline
```

# Veri seti

```python
np.random.seed(42)
df_size = 20
df = pd.DataFrame({
    'Kategori': np.random.choice(['Bilgisayar','Tablet','Telefon'], size=df_size),
    'Değer': np.random.normal(loc=10, scale=3, size=df_size),
    'Zaman': np.random.randint(1,10, size=df_size),
})
```

# Grafikler

---

## Çizgi grafiği

```python
plt.figure(figsize=(6,4))
sns.lineplot(x='Değer', y='Zaman', data=df, color='purple', marker="o")
plt.title("Çizgi grafiği (Seaborn)")
plt.show()
```

## Dağılım grafiği

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='Değer', y='Zaman', data=df, color='darkred', s=200)
plt.title("Dağılım grafiği (Seaborn)")
plt.show()
```

## Bar grafiği

```python
sns.set_style("whitegrid")
sns.set_context("talk")

plt.figure(figsize=(8,6))
bar = sns.barplot(
    x='Zaman',
    y='Değer',
    data=df,
    edgecolor='black',
    saturation= 0.8,
    palette='pastel'
)

#plt.xticks(rotation=45, ha="right")
plt.title("Barplot Seaborn")
plt.tight_layout()
plt.show()
```

## Boxplot grafiği

```python
plt.figure(figsize=(6,4))
sns.boxplot(
    x='Zaman',
    y='Değer',
    data=df,
    palette="Set1"
)

plt.title("Boxplot Seaborn")
plt.show()
```

## Violin grafiği

```python
sns.set_context('notebook')

plt.figure(figsize=(6,4))
sns.violinplot(
    x='Zaman',
    y='Değer',
    data=df,
    palette="Set1"
)

plt.title("Boxplot Seaborn")
plt.show()
```

## Eşleştirme grafiği

```python
sns.pairplot(df, hue="Kategori")
plt.title("Pairplot örneği")
plt.show()
```
