```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

## Çeşitli grafik özelleştirmeleri

```python
plt.figure(figsize=(6,4))
plt.plot(x_veri, y_veri, marker="o", markersize=8, color="darkorange", linestyle="--", linewidth=5)
plt.title("Deneme grafiği", fontsize = 16)
plt.xlabel("X ekseni")
plt.ylabel("Y ekseni")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.plot(x_veri, y_veri, marker="o", color="indigo", linestyle=":")
plt.title("Deneme grafiği", fontsize = 16)
plt.xlabel("X ekseni")
plt.ylabel("Y ekseni")
plt.tight_layout()
plt.grid()
plt.show()
```
