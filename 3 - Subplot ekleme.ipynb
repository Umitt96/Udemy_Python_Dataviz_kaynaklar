```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

## Veri seti

```python
x_veri = np.arange(1,7)
y_veri = np.array([5,2,6,8,4,3])
etiketler = ['A','B','C','D','E','F']
```

```python
# tekli veri seti
np.random.seed(1)
x_tek_veri = 4 + np.random.normal(0, 1.5, 200)
```

## Subplot (Ã‡oklu grafik ekleme)

```python
fig, axs = plt.subplots(1,2, figsize=(14,6))
axs[0].plot(x_veri,y_veri, color="darkred", marker="o")
axs[0].set_title("Ã‡izgi grafiÄŸi")

axs[1].bar(x_veri,y_veri, color="dodgerblue")
axs[1].set_title("Bar grafiÄŸi")

plt.show()
```

```python
fig, axs = plt.subplots(2,1, figsize=(3,7))
axs[0].plot(x_veri,y_veri, color="darkred", marker="o")
axs[0].set_title("Ã‡izgi grafiÄŸi")
axs[0].set_ylabel("SayÄ±lar")

axs[1].hist(x_tek_veri, color="dodgerblue")
axs[1].set_title("Histgoram grafiÄŸi")
axs[1].set_ylabel("Miktar")

plt.show()
```
