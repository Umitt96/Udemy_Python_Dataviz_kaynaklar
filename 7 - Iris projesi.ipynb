# Veri Görselleştirme

## Dağılım grafikleri

```python
sns.pairplot(iris, hue="species", corner=True, palette="viridis")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(data=iris, x="petal_length", hue="species", edgecolor="black", linewidth=1.5, palette="viridis")
plt.title("Çanak Yaprak Uzunluk Dağılımı (petal_length)")
plt.show()
```

## Sayısal ve Kategorik veriler

```python
plt.figure(figsize=(8,6))
sns.boxplot(data = iris, x = "species", y = "sepal_length", palette="viridis")
sns.stripplot(data = iris, x = "species", y = "sepal_length", color="black", alpha=0.7)
plt.title("Türlere Göre Taç Yaprak Uzunluğu", fontsize=16)
plt.xlabel("Çiçek Türleri")
plt.ylabel("Taç Yaprak Uzunluğu")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(data=iris, x="species", y="petal_width", palette="viridis", capsize=0.2)
plt.title("Çanak Yaprak Genişlikleri", fontsize=16)
plt.tight_layout()
plt.show()
```

## Isı Haritası (Heatmap)

```python
plt.figure(figsize=(6,6))
corr = iris.drop(columns="species").corr()

sns.heatmap(corr, annot=True, cmap="viridis_r", square=True, linewidths=3)
plt.title("Iris Korelasyon Grafiği", fontsize=16)
plt.show()
```
